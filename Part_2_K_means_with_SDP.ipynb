# 2. k-means Clustering with Semidefinite Programming - 45 points

Clustering is and unsupervised machine learning problem in which we try to partition a given dataset into $k$ subsets based on distance between data points or similarity among them. The goal is to find $k$ centers and to assign each data point to one of the centers such that the sum of the square distances between them are minimal [1]. This problem is known to be NP-hard.

## Clustering problem
Given a set of $n$ points in a $d-$dimensional Euclidean space, denoted by
\begin{equation*}
S = \{ \mathbf{s}_i = (s_{i1}, \cdots, s_{id})^\top~\in \mathbb{R}^d ~~ i = 1, \cdots, n\}
\end{equation*}
find an assignment of the $n$ points into $k$ disjoint clusters $\mathcal{S} = (S_1, \cdots, S_k)$ whose centers are $\mathbf{c}_j(j = 1, \cdots, k)$ based on the total sum of squared Euclidean distances from each point $\mathbf{s}_i$ to its assigned cluster centroid $\mathbf{c}_i$, i.e.,
\begin{equation}
f(S,\mathcal{S}) = \sum_{j=1}^k\sum_{i=1}^{|S_j|}\|\mathbf{s}_i^{j} - \mathbf{c}_j \|^2,
\label{eq:cluster}\tag{Clustering Problem}
\end{equation}
where $|S_j|$ is the number of points in $S_j$, and $\mathbf{s}_i^{j} $ is the $i^{th}$ point in $S_j$.

## Lloyd's algorithm for k-means



  **1.** Choose initial cluster centers $\mathbf{c}_1, \mathbf{c}_2, \cdots , \mathbf{c}_k$
  
  **2.** Repeat until convergence:
  $$\begin{cases}
  \text{Assignment step: }~~  \mathbf{s}_i ~~\text{belongs to cluster } j\text{, where} j := \text{argmin}_{j \in [1,k]} \|\mathbf{s}_i-\mathbf{c}_j\|   \\
  \text{Update each cluster center: }~~~~~~~~ \mathbf{c}_j = \frac 1{|S_j|} \sum_{i=1}^{|S_j|} \mathbf{s}_i^{j} \\
  \end{cases}$$


Note that the algorithm converges to local optimal points, so
\eqref{eq:cluster} can be arbitrarily bad depending on the
initialization of the cluster centers.


 **SDP relaxation of the problem:** The work (Peng & Wei, 2007)
proposes an SDP-relaxation to approximately solve the aforementioned
model-free $k-$means clustering problem. The resulting optimization
problem (See section (2) of (Peng & Wei, 2007) for details of this
    relaxation and Lecture 14 for a brief introduction.) takes the standard semidefinite programming form

$$\mathbf{X}^\star \in \arg \min_{ \mathbf{X} } \bigg\{ \langle \boldsymbol{C}, \mathbf{X} \rangle : ~\underbrace{\mathbf{X} \mathbf{1} =  \mathbf{1}}_{A_1(\mathbf{X}) = b_1},\underbrace{~\mathbf{X}^{\top} \mathbf{1} = \mathbf{1}}_{A_2(\mathbf{X}) = b_2},~\underbrace{\mathbf{X} \geq 0}_{B(\mathbf{X}) \in \mathcal{K}},~\underbrace{\mathrm{Tr}(\mathbf{X}) \leq \kappa, ~\mathbf{X} \in \mathbb{R}^{p\times p} ,~\mathbf{X}\succeq 0 }_{\mathcal{X}} \bigg\},
\label{prob1}\tag{Problem 1}$$
where $C \in \mathbb{R}^{p \times p}$ is the
Euclidean distance matrix between the data points.
$\mathrm{Tr}(\mathbf{X}) \leq \kappa$ enforces approximately low-rank
solutions, the linear inclusion constraint $\mathbf{X} \geq 0$ is element-wise
nonnegativity of ${\mathbf{X}}$, the linear equality constraints
$\mathbf{X}\mathbf{1} = \mathbf{1}$ and $\mathbf{X}^{\top} \mathbf{1} = \mathbf{1}$
require row and column sums of $\mathbf{X}$ to be equal to 1's, and
$\mathbf{X} \succeq 0$ means that $\mathbf{X}$ is positive semi-definite. Recall that
$\mathrm{Tr}(\mathbf{X}) = \|\mathbf{X}\|_\ast$ for any positive semi-definite matrix
$\mathbf{X}$.

The SDP in Problem 1 can be reformulated as
$$
\min_{\mathbf{X} \in \mathcal{X}}~~~ f(\mathbf{X}) + g_1(A_1(\mathbf{X})) + g_2(A_2(\mathbf{X})) \qquad\text{subject to}\qquad B(\mathbf{X}) \in \mathcal{K}, \label{prob2}\tag{Equation 2}$$
where $f(\mathbf{X}) = \langle  \boldsymbol{C}, \mathbf{X} \rangle$ is a
smooth convex function, $g_1=\delta_{\{b_1\}}(\cdot)$ is the indicator
function of singleton $\{b_1\}$, $g_2=\delta_{\{b_2\}}(\cdot)$ is the
indicator function of singleton $\{b_2\}$ and ${\mathcal{K}}$ is the
positive orthant for which computing the projection is easy.

### Question 2.1.1

Frank–Wolfe / CGM would be valid **only if** we can compute the linear minimization oracle (LMO)
$$\arg\min_{S\in\mathcal{D}} \langle G, S\rangle$$
over the feasible set $\mathcal{D}$.

In our problem, the feasible set is the intersection of several constraints (PSD + trace bound + affine equalities + nonnegativity). The course slides explicitly say that **classical CGM does not apply** when we have something like

So, “vanilla” Frank–Wolfe is not the practical choice here. The fix (also in the slides) is to keep an easy domain $X$ for the LMO and move the hard constraint into a **quadratic penalty**, then increase the penalty parameter (homotopy), which gives HCGM.

### Question 2.1.2

We need to show that
$$\mathcal{X}=\{X\in\mathbb{R}^{p\times p}:\ X\succeq 0,\ \mathrm{Tr}(X)\le \kappa\}$$
is convex.

Take any $X,Y\in\mathcal{X}$ and any $\theta\in[0,1]$. Consider
$$Z=\theta X+(1-\theta)Y.$$

1) **PSD part:** The set of PSD matrices is convex, so if $X\succeq 0$ and $Y\succeq 0$, then
$$Z=\theta X+(1-\theta)Y \succeq 0.$$

2) **Trace part:** Trace is linear, so
$$\mathrm{Tr}(Z)=\mathrm{Tr}(\theta X+(1-\theta)Y)=\theta\,\mathrm{Tr}(X)+(1-\theta)\,\mathrm{Tr}(Y).$$
Since $\mathrm{Tr}(X)\le \kappa$ and $\mathrm{Tr}(Y)\le \kappa$,
$$\mathrm{Tr}(Z)\le \theta \kappa+(1-\theta)\kappa=\kappa.$$

Both conditions hold, so $Z\in\mathcal{X}$. Therefore $\mathcal{X}$ is convex by the definition of convexity.

## 1) Gradient of the linear term

Let
$$
\sum_{i,j} C_{ij}X_{ij}.
$$
Then for each entry,
$$
\frac{\partial}{\partial X_{k\ell}}\sum_{i,j} C_{ij}X_{ij}=C_{k\ell}.
$$
So
$$
\nabla\left(\sum_{i,j} C_{ij}X_{ij}\right)=C.
$$

### 2) Gradient of the row-sum penalty

Start from
$$
\phi_{\text{row}}(X)=\frac{\beta}{2}\|X\mathbf{1}-\mathbf{1}\|^2.
$$
Write the squared norm as a sum of coordinates. The $i$-th entry of $X\mathbf{1}$ is the sum of row $i$:
$$
(X\mathbf{1})_i=\sum_{j=1}^p X_{ij}.
$$
So
$$
\|X\mathbf{1}-\mathbf{1}\|^2
=
\sum_{i=1}^p\left((X\mathbf{1})_i-1\right)^2
=
\sum_{i=1}^p\left(\sum_{j=1}^p X_{ij}-1\right)^2,
$$
and therefore
$$
\phi_{\text{row}}(X)=\frac{\beta}{2}\sum_{i=1}^p\left(\sum_{j=1}^p X_{ij}-1\right)^2.
$$

Now take the partial derivative with respect to one entry $X_{k\ell}$.

Only the term with $i=k$ involves $X_{k\ell}$, so
$$
\frac{\partial \phi_{\text{row}}}{\partial X_{k\ell}}
=
\frac{\beta}{2}\,
\frac{\partial}{\partial X_{k\ell}}
\left(\sum_{j=1}^p X_{kj}-1\right)^2.
$$
Differentiate:
$$
\frac{\partial \phi_{\text{row}}}{\partial X_{k\ell}}
=
\beta\left(\sum_{j=1}^p X_{kj}-1\right),
$$
because $\frac{\partial}{\partial X_{k\ell}}\left(\sum_{j=1}^p X_{kj}-1\right)=1$.

So the gradient entry at position $(k,\ell)$ is
$$
\big(\nabla \phi_{\text{row}}(X)\big)_{k\ell}
=
\beta\left(\sum_{j=1}^p X_{kj}-1\right).
$$

So row $k$ of the gradient looks like
$$
\Big[
\beta(\sum_j X_{kj}-1),\
\beta(\sum_j X_{kj}-1),\
\dots,\
\beta(\sum_j X_{kj}-1)
\Big].
$$
That is exactly
$$
\beta\left(\sum_{j=1}^p X_{kj}-1\right)\mathbf{1}^\top.
$$

Now define the vector of row residuals:
$$
X\mathbf{1}-\mathbf{1}
=
\begin{bmatrix}
\sum_j X_{1j}-1\\
\sum_j X_{2j}-1\\
\vdots\\
\sum_j X_{pj}-1
\end{bmatrix}.
$$
The $k$-th entry of this vector is exactly $\sum_j X_{kj}-1$.

When we multiply this vector by $\mathbf{1}^\top$, we replicate the $k$-th entry across all columns in row $k$:
$$
\big((X\mathbf{1}-\mathbf{1})\mathbf{1}^\top\big)_{k\ell}
=
(X\mathbf{1}-\mathbf{1})_k\cdot 1
=
\sum_{j=1}^p X_{kj}-1.
$$

Comparing with the entrywise gradient formula above, we get
$$
\nabla\left(\frac{\beta}{2}\|X\mathbf{1}-\mathbf{1}\|^2\right)
=
\beta\,(X\mathbf{1}-\mathbf{1})\mathbf{1}^\top.
$$

---

### 3) Gradient of the column-sum penalty

Start from
$$
\phi_{\text{col}}(X)=\frac{\beta}{2}\|X^\top\mathbf{1}-\mathbf{1}\|^2.
$$
The $\ell$-th entry of $X^\top\mathbf{1}$ is the sum of column $\ell$:
$$
(X^\top\mathbf{1})_\ell=\sum_{k=1}^p X_{k\ell}.
$$
So
$$
\|X^\top\mathbf{1}-\mathbf{1}\|^2
=
\sum_{\ell=1}^p\left((X^\top\mathbf{1})_\ell-1\right)^2
=
\sum_{\ell=1}^p\left(\sum_{k=1}^p X_{k\ell}-1\right)^2,
$$
and therefore
$$
\phi_{\text{col}}(X)=\frac{\beta}{2}\sum_{\ell=1}^p\left(\sum_{k=1}^p X_{k\ell}-1\right)^2.
$$

Now differentiate with respect to $X_{k\ell}$.

Only the term for that specific column index $\ell$ contains $X_{k\ell}$, so
$$
\frac{\partial \phi_{\text{col}}}{\partial X_{k\ell}}
=
\frac{\beta}{2}\,
\frac{\partial}{\partial X_{k\ell}}
\left(\sum_{i=1}^p X_{i\ell}-1\right)^2.
$$
Differentiate:
$$
\frac{\partial \phi_{\text{col}}}{\partial X_{k\ell}}
=
\beta\left(\sum_{i=1}^p X_{i\ell}-1\right),
$$
because $\frac{\partial}{\partial X_{k\ell}}\left(\sum_{i=1}^p X_{i\ell}-1\right)=1$.

So entrywise,
$$
\big(\nabla \phi_{\text{col}}(X)\big)_{k\ell}
=
\beta\left(\sum_{i=1}^p X_{i\ell}-1\right).
$$


So column $\ell$ of the gradient looks like
$$
\beta\left(\sum_{i=1}^p X_{i\ell}-1\right)\mathbf{1}.
$$

The vector of column residuals is
$$
X^\top\mathbf{1}-\mathbf{1}
=
\begin{bmatrix}
\sum_i X_{i1}-1\\
\sum_i X_{i2}-1\\
\vdots\\
\sum_i X_{ip}-1
\end{bmatrix}.
$$
Its $\ell$-th entry is $\sum_i X_{i\ell}-1$.

Multiplying $\mathbf{1}(X^\top\mathbf{1}-\mathbf{1})^\top$ replicates that residual down the entire column:
$$
\big(\mathbf{1}(X^\top\mathbf{1}-\mathbf{1})^\top\big)_{k\ell}
=
1\cdot (X^\top\mathbf{1}-\mathbf{1})_\ell
=
\sum_{i=1}^p X_{i\ell}-1.
$$

Thus, matching entries gives
$$
\nabla\left(\frac{\beta}{2}\|X^\top\mathbf{1}-\mathbf{1}\|^2\right)
=
\beta\,\mathbf{1}(X^\top\mathbf{1}-\mathbf{1})^\top.
$$

---

## 4) Gradient of the cone-distance penalty

From Step 0,
$$
\mathrm{dist}^2(X,\mathcal{K})=\|X-\max(X,0)\|_F^2
=\sum_{i,j}\left(X_{ij}-\max(X_{ij},0)\right)^2.
$$
So the cone penalty is
$$
\frac{\beta}{2}\sum_{i,j}\left(X_{ij}-\max(X_{ij},0)\right)^2.
$$

Using the Danskin hint, the gradient is
$$
\nabla\left(\frac{\beta}{2}\|X-\max(X,0)\|_F^2\right)
=
\beta\,(X-\max(X,0)).
$$

---

## 5) Final gradient

Combine all parts:
$$
\nabla F_\beta(X)
=
C
+\beta\,(X\mathbf{1}-\mathbf{1})\mathbf{1}^\top
+\beta\,\mathbf{1}(X^\top\mathbf{1}-\mathbf{1})^\top
+\beta\,(X-\max(X,0)).
$$

  #### Homotopy Conditional Gradient Method (HCGM)
  ---------------------------------------------------------------------------------------------------------------------------------
  **1.** Choose $x^0 \in\mathcal{X}$ and $\beta_0 >  0$
  
  **2.** For $k=1, 2, \ldots$ perform:
  $$
  \begin{cases}
  \gamma_k &= 2/(k+1), ~~\text{and}~~ \beta_k = \beta_0 \sqrt{k+1} \\
  v_k &=  \nabla f(x_k) + \beta_k [A_1^\top (A_1(x_k) - b_1) + A_2^\top (A_2(x_k) - b_2) + (x_k - \text{proj}_{\mathcal{K}}(x_k))]\\
  \hat{x}^k &:= \mathrm{argmin}_{x \in \mathcal{X}} \left \langle  v_k,  x \right \rangle \\
  x^{k+1} &:= (1-\gamma_k)x^k + \gamma_k\hat{x}^k
  \end{cases}
  $$
  
  **3.** Output:$~x^{k+1}$
  
  ---------------------------------------------------------------------------------------------------------------------------------

HCGM is designed for solving constrained minimization problems of the form in (Equation 2). We define such a constrained problem in the following cell, where we specify an objective $f$ and a set of penalities $\{g_i\}$.


````python
proj_K = lambda X: np.maximum(X, 0.0)

f = Function(
    f = lambda X : np.sum(C.flatten()*X.flatten()),
    grad = lambda X: C
)

g1 = Function(
    f = lambda X : 0.5 * np.linalg.norm((A1(X) - b))**2,
    grad = lambda X: At1(A1(X) - b)
)

g2 = Function(
    f = lambda X : 0.5 * np.linalg.norm((A2(X) - b))**2,
    grad = lambda X: At2(A2(X) - b)
)

g3 = Function(
    f = lambda X : 0.5 * np.linalg.norm(X - proj_K(X), ord="fro")**2,
    grad = lambda X: 1000 * (X - proj_K(X)) # <-- fill in (x_k - proj_K(x_k))
)

p = ConstrainedProblem(f = f, penalties = [g1, g2, g3])
````


**Remark**: The 1000 factor in the term for $(x_k - \text{proj}_{\mathcal{K}}(x_k))$ in Algorithm 1 is in order to obtain a better practical convergence. This basically corresponds to having different penalty parameters for different constraints.

#### Question 2.2.1 (5 pts)

Complete the missing lines in the function
definitions of `HCGM`, which implements Homotopy CGM.

````python
@dataclass
class HCGM_state(OptState):
    x_k: Matrix
    k: int
    beta0: float
````

````python
def HCGM_initialize(f, Xzero, beta0):
    k = 0
    return HCGM_state(x_k = Xzero, k = k, beta0 = beta0)
````

For the update recall that a `ConstrainedProblem` can be unpacked as follows:

- `f` which is a `Function` that has a `grad` method
- `penalties` which is a list of functions each having a `grad` method

````python
def HCGM_update(p : ConstrainedProblem, state : HCGM_state):
    f, penalties = p
    x_k, k, beta0 = state

    gamma_k = 2.0 / (k + 2.0)
    beta_k = beta0 * np.sqrt(k + 1.0)

    v_k = f.grad(x_k) + beta_k * sum(pen.grad(x_k) for pen in penalties)

    X_bar = lmo(v_k, kappa)

    X_next = (1.0 - gamma_k) * x_k + gamma_k * X_bar

    return HCGM_state(x_k = X_next, k = k+1, beta0 = beta0)
````

````python
HCGM = OptAlgorithm(name="HCGM", init_state = HCGM_initialize, state_update = HCGM_update)
````

### Algorithm 2. PD3O splitting

To implement PD3O we need to specify the proximal operator of $h$ and $g^*$ as well as the linear operator $A$. We do so in the following cell. Note that the `DualVariable` class is created to keep track of all three dual variable $(y_1,y_2,y_3)$ in a concise way.
In the following cells fill in the proximal operators of $h$ and $g$.

**Hint**: Recall that the proximal operator of an indicator function is a projection. We give you the `projSDP`, which projects onto $\mathcal X$. Further, the proximal operator of a seperable sum is a concatenation of the proximal operators of each term (see the lecture on proximal methods).

````python
def ispsd(X, tol=1e-10):
    Xs = 0.5 * (X + X.T)
    return np.all(np.linalg.eigvalsh(Xs) >= -tol)

f = Function(
    f = lambda X : np.sum(C.flatten()*X.flatten()),
    grad = lambda X : C
)

h = Function(
    f = lambda X : 0 if (ispsd(X) and np.trace(X) <= kappa) else float('inf') ,
    prox = lambda tau, X: projSDP(0.5 * (X + X.T), kappa)
)

g = Function(
    f = lambda y: 0 if (y.y_1==b) and (y.y_2==b) and (np.all(y.y_3 >=0)) else float('inf'),
    prox = lambda sigma, y: DualVariable(b, b, np.maximum(y.y_3, 0.0))
)

def A(x):
    return DualVariable(A1(x), A2(x), x)

def A_T(y):
    return At1(y.y_1)+ At2(y.y_2)+ y.y_3

composite = (f, g, h, A, A_T)
````

#### Question 2.2.3 (5 pts)

Complete the missing lines in the function
definitions of `PD3O`

**Remarks:**

- In our case, we advise you to **multiply the step-size for $y_3$ by $10^4$** to obtain a better practical convergence. It is possible to use different dual step sizes, $\{ \sigma_1 , \sigma_2, \sigma_3 \}$.  Such a diagonal variable stepsize is supported through `[sigma1, sigma2, sigma3]*dual_variable`.
- The algorithm needs the proximal operator of $g^*$, the Fenchel dual of $g$, using the Moreau identity express the prox of $g^*$ in terms of the prox of $g$. You will need to multiply by the stepsize `Sigma` and its inverse `Sigma_inv` which we provide in the code below.

````python
@dataclass
class PD3O_state(OptState):
    x_k: Matrix
    y_k: DualVariable(Vector, Vector, Matrix)
    tau: float
    sigma: float
````

````python
def PD3O_initialize(composite, x_zero):
    L = 1e2
    tau = 1/L
    return PD3O_state(x_k = x_zero, y_k = A(x_zero), tau=tau, sigma=1/(L**2*tau))

def PD3O_update(composite, state):
    f, g, h, A, A_T = composite
    y_k = state.y_k
    x_k, _, tau, sigma = state

    next_x_k = h.prox(tau, x_k - tau * (f.grad(x_k) + A_T(y_k)))

    x_bar = 2.0 * next_x_k - x_k + tau * f.grad(x_k) - tau * f.grad(next_x_k)

    Sigma = [sigma, sigma, sigma*1e4]
    Sigma_inv = [1/sigma, 1/sigma, 1/(sigma*1e4)]

    proj = g.prox(Sigma_inv, Sigma_inv * y_k + A(x_bar))
    next_y_k = y_k + Sigma * (A(x_bar) - proj)

    return PD3O_state(next_x_k, next_y_k, tau, sigma)



````

````python
PD3O = OptAlgorithm(name="PD3O", init_state = PD3O_initialize, state_update = PD3O_update)
````

<font color=‘blue’>
    
The provided plot compares the convergence behavior of HCGM and PD3O in terms of the relative objective residual 
$\big(f(X)-f^\star\big)/|f^\star|$ as a function of wall-clock time. 
Both algorithms eventually reduce the objective residual, but they exhibit different convergence patterns. 
PD3O shows a faster decrease in the objective value once feasibility is sufficiently enforced, leading to a rapid drop in the residual after an initial transient phase. 
In contrast, HCGM decreases the objective more gradually, which is consistent with its homotopy strategy: during early iterations the penalty parameter is small and the algorithm prioritizes objective minimization over strict constraint enforcement, while feasibility is enforced more strongly only as the penalty increases.

The final objective values attained by both methods are slightly below the reference optimal value $opt_{val}$. 
This does not contradict optimality, since $opt_{val}$ corresponds to the minimum of the objective over the feasible SDP set, whereas the plotted objective values are evaluated at iterates that are not yet exactly feasible. 
Because both HCGM and PD3O temporarily explore infeasible points, the linear objective can attain values lower than the constrained optimum. 
As the feasibility residuals decay to near-zero, the objective values approach $opt_{val}$ up to numerical tolerance.
